```python
pdf_text=[]
```

```python
f = open('/content/PREM_RESUME.pdf', 'rb')
pdf_reader = PyPDF2.PdfReader(f)
for p in range(len(pdf_reader.pages)):
  page = pdf_reader.pages[p]
  pdf_text.append(page.extract_text())
f.close()
```

```python
pdf_text
```

```python
len(pdf_text)
```

```python
for page in pdf_text:
  print(page)
  print('\n')
```

```python
with open('extracted_text.txt', 'w') as f:
    for page in pdf_text:
        f.write(page)
        f.write('\n')
```

Regular expressions :

Regular expressions (regex) are patterns used to match, search, and manipulate text based on specific rules. They help find strings that fit a format, like emails, phone numbers, or dates, using special characters and symbols.
