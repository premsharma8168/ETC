Regular Expression

Regular expressions or RegEx is a sequence of characters mainly used to find or replace patterns embedded in the text.

re.match( )

The re.match function returns a match object on success and none on failure

```python
import re

#match a word at the beginning of a string
result = re.match('The',r'The cat jumped gracefully onto the windowsill, chasing a beam of sunlight.') #change it with 'The'
print(result)
```

```python
print(result.group())
```

re.search( )

Matches the first occurrence of a pattern in the entire string(and not just at the beginning)

```python
result = re.search('bright',r'The sun is bright. She wore bright colors.')
print(result.group())
```

re.findall( )

It will return all the occurrences of the pattern from the string

```python
result = re.findall('bright',r'The sun is bright. She wore bright colors.')
print(result)
```

Metacharacters

characters with special meaning

1. (.)  matches any character (except newline character)

```python
str = r'The cat jumped gracefully onto the windowsill, chasing a beam of sunlight.'
```

```python
x = re.findall("ju.", str)
x2 = re.findall("win...", str)

print(x)
print(x2)
```

2. (^) starts with

It checks whether the string starts with the given pattern or not.

```python
x = re.findall("^The", str)

if (x):
  print("Yes")
else:
  print("No match")
```

3. ($) ends with

It checks whether the string ends with the given pattern or not.

```python
str="Today is a very good weather"
```

```python
x = re.findall("weather$", str)

if (x):
  print("Yes")
else:
  print("No match")
```

4. (*) matches for zero or more occurrences of the pattern to the left of it

```python
str1 = "easy easssy eay ey"
```

```python
x = re.findall("eas*y", str1)
print(x)

if (x):
  print("Yes")
else:
  print("No match")
```

5. (+) matches one or more occurrences of the pattern to the left of it

```python
x = re.findall("eas+y", str1)
print(x)

if (x):
  print("Yes")
else:
  print("No match")
```

6. (?) matches zero or one occurrence of the pattern left to it.

```python
x = re.findall("eas?y",str1)

print(x)

if (x):
  print("Yes")
else:
  print("No match")
```

7. (|) either or

```python
x = re.findall("very|good", str)
print(x)

if (x):
  print("Yes")
else:
  print("No match")
```

8. @\w+ ; checks whether '@' is present followed by word characters

9. \b[A-Z][a-z]+\b : finds words starting with a capital letter followed by lowercase letters

10. \d+(\.\d+)? : matches integers or decimals

```python
x = re.findall(r"\d+(\.\d+)?", str)
print(x)

if x:
    print("Yes")
else:
    print("No match")
```

11. [A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,} : helps to find the email in the string

```python
x = re.findall(r" [A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", str)
print(x)

if x:
    print("Yes")
else:
    print("No match")
```

12. https?://\S+|www\.\S+ : finds the link in a string

```python
x = re.findall(r"https?://\S+|www\.\S+", str)
print(x)

if x:
    print("Yes")
else:
    print("No match")
```

13. [$₹€]\d+(\.\d{2})? : matches currency amounts

```python
x = re.findall(r"[$₹€]\d+(\.\d{2})?", str)
print(x)

if x:
    print("Yes")
else:
    print("No match")
```

14. "([^"]+)" : matches text inside double quotes.

```python
x = re.findall(r'([^"]+)', str)
print(x)

if (x):
  print("Yes")
else:
  print("No match")
```
